```python
# import required libraries
import pandas as pd
import numpy as np
```

```python
# tabular data
data = {
    "where": ["Home", "Home","Away", "Home", "Away","Away","Home","Home","Away","Home"],
    "when": ["7pm", "7pm", "7pm", "5pm", "9pm", "7pm","7pm", "7pm","7pm","9pm"],
    "Fred_Starts": ["Yes", "Yes", "Yes", "No", "Yes", "No","No","Yes","Yes","No"],
    "Joe_Offense": ["Center", "Forward","Forward", "Forward","Forward","Center", "Forward","Center","Center","Forward"],
    "Joe_Defense": ["Forward", "Center", "Forward", "Center", "Forward", "Forward", "Center","Center","Center","Center"],
    "Opp_C": ["Tall", "Short","Tall", "Tall", "Short","Tall","Tall","Tall","Short","Short"],
    "outcome": ["Won", "Won", "Won", "Lost", "Lost", "Won","Lost", "Won","Won", "Lost"],

}
#output data
pd.DataFrame(data)
```

```python
#Entropy  = Probalility(log(2)*Probability)
#Information gain = Outcome entropy - attributes entropy

#function to determining the best attribute

def bestAttribute(outcome, attribute,attribute_name):

  #find outcome entropy

  #return the probabilty of win and loss in the outcome column
  prob_outcome = pd.Series(data['outcome']).value_counts(normalize = True)

  #add entropy(win + lost)
  outcome_entropy = np.sum(-(prob_outcome * np.log2(prob_outcome)))

  #create a dataframe for attribute and outcome
  attribute_outcome_df = pd.DataFrame(
      {
          "attribute": attribute,
          "outcome": outcome
      }
  )
  #list unique categories in the attributes selected
  values = list(attribute_outcome_df['attribute'].unique())
  #print("values are ", values)


  #print the length of the data in attributes
  total = len(attribute_outcome_df['attribute'])
  #print("values are ", total)

  attribute_entropy = 0

  for val in values:

    #get the total of a given category of attribute
    count = attribute_outcome_df[attribute_outcome_df['attribute'] == val]['outcome'].value_counts().sum()
    #print(count)

    #get the probalities of win/loss for each individual category of attribute
    probs = attribute_outcome_df[attribute_outcome_df['attribute'] == val]['outcome'].value_counts(normalize = True)
    #print(probs)

    #entropy of an attribute
    attribute_entropy += np.sum(-probs *np.log2(probs)) * (count/total)

  information_gain = outcome_entropy - attribute_entropy


  print("Attribute: {} \nEntropy: {:.4f} \nInformation Gain: {:.4f}\n\n\n".format(attribute_name, attribute_entropy, information_gain))

  return attribute_entropy, information_gain
```

```python
bestAttribute(data['outcome'], data['when'],"when")
```

```python
# Finding the most useful attribute

most_useful_attribute = -np.inf 
attribute1 = ""

for key, value in data.items():
  if key == "outcome":
    continue

  ent, inf = bestAttribute(data["outcome"], data[key], key)

  if inf > most_useful_attribute:
    most_useful_attribute = inf
    attribute1 = key

print(f"\n Most useful Attribute is {attribute1}  with information gain {most_useful_attribute}")
```
